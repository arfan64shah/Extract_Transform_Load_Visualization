```python
#import pandas library 
import pandas as pd

#load and read the data using pandas
dataset = pd.read_csv('avocado.csv')

#display some rows
dataset.head()
```

```python
#2: display the column names
dataset.columns
```

```python
#3: get the average value of large bags
#we will select the large bags columns and then use mean() function to get average of the whole column
dataset['Large Bags'].mean()
```

```python
#4: best 5 records for sale price/average price
dataset.nlargest(5, 'AveragePrice')
```

```python
#5: best year in terms of production
yearData= dataset.nlargest(1, 'Total Volume')
yearData['Date']
```

```python
#6: total bags should be equal to small and large bags
total_bags = dataset['Total Bags'].sum()

print(total_bags)
small_bags = dataset['Small Bags'].sum()
large_bags = dataset['Large Bags'].sum()
xlarge_bags = dataset['XLarge Bags'].sum()
print("small bags: ", small_bags)
print("large bags: ", large_bags)
print("extra large bags: ", xlarge_bags)

bags = small_bags + large_bags + xlarge_bags
print(bags)
```

```python
#since the bags are not equal, still we can use a different method
if(total_bags == bags):
    print("Bags are equal")
else:
    print("Bags are not equal")
```
