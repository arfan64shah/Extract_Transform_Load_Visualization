```python
#import library and read data
import pandas as pd
dataset = pd.read_csv('credit_dataset.csv')
dataset.head()
```

```python
#descriptive statistics
dataset.describe()
```

So there are no zero values

```python
#find all the columns
dataset.columns
```

```python
#using get dummies function to deal with object type columns
new_dataset = pd.get_dummies(dataset, prefix=['GENDER', 'CAR', 'REALITY', 'INCOME_TYPE', 'EDUCATION_TYPE', 'FAMILY_TYPE', 'HOUSE_TYPE'])
#final_dataset = pd.concat((dataset, new_dataset), axis=1)
#final_dataset = new_dataset.drop(['GENDER', 'CAR', 'REALITY', 'INCOME_TYPE', 'EDUCATION_TYPE', 'FAMILY_TYPE', 'HOUSE_TYPE'], axis=1)
new_dataset.head()
```

```python
print(new_dataset.columns)
print(new_dataset.shape)
```

So the columns have been increased from 20 to 40 after applying get_dummies function

```python
#now find again the datatypes
print(new_dataset.dtypes)
print(len(new_dataset.columns))
```

so there are no objects and it is almost ready to apply any model

```python
#as the target column is in between and we need to remove it and add it to the end
move_column = new_dataset.pop("TARGET")
new_dataset.insert(39, "TARGET", move_column)
new_dataset.head()
```

```python
#As the data is ready and we split it into dependent and independent variables
x = new_dataset.iloc[:, :-1]
y = new_dataset.iloc[:, -1]
```

```python
#now split data into train and test
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 0)
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
#now lets apply random forest
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier()
classifier.fit(x_train, y_train)
y_pred = classifier.predict(x_test)
```

```python
#now find accuracy metrices
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score, classification_report
print("Accuracy is", accuracy_score(y_test, y_pred))
print("F1 score is ", f1_score(y_test, y_pred))
print("Precision score is ", precision_score(y_test, y_pred))
print("Recall score is ", recall_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
#now find confusion matrix and see
from sklearn.metrics import confusion_matrix
con_matrix = confusion_matrix(y_test, y_pred)
print(con_matrix)

#plot confusion matrix
sb.heatmap(con_matrix, annot=True)
```

so we get this result and this result is not very good. We will try using undersampling and oversampling

```python
#oversampling
from imblearn.over_sampling import SMOTE

smote = SMOTE(random_state=1)
over_x_train, over_y_train = smote.fit_resample(x_train, y_train)

#find the number of instances in y train
print(over_y_train.value_counts())

#now again apply random forest using the resampled data
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier(criterion = 'entropy', random_state = 1)
classifier.fit(over_x_train, over_y_train)
y_predict = classifier.predict(x_test)

#now find the metrices
print("Accuracy after resampling is ", accuracy_score(y_test, y_predict))
print("Recall after resampling is ", recall_score(y_test, y_predict))
print("Precision after resampling is ", precision_score(y_test, y_predict))
print("F1 score after oversampling is ", f1_score(y_test, y_predict))
print(classification_report(y_test, y_predict))
```

we can see the report is improved

```python
#now lets do under sampling
#oversampling
from imblearn.under_sampling import RandomUnderSampler

undersampler = RandomUnderSampler(random_state=1)
under_x_train, under_y_train = undersampler.fit_resample(x_train, y_train)

#find the number of instances in y train
print(under_y_train.value_counts())

#now again apply random forest using the resampled data
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier(criterion = 'entropy', random_state = 1)
classifier.fit(under_x_train, under_y_train)
y_predict = classifier.predict(x_test)

#now find the metrices
print("Accuracy after resampling is ", accuracy_score(y_test, y_predict))
print("Recall after resampling is ", recall_score(y_test, y_predict))
print("Precision after resampling is ", precision_score(y_test, y_predict))
print("F1 score after oversampling is ", f1_score(y_test, y_predict))
print(classification_report(y_test, y_predict))
```
