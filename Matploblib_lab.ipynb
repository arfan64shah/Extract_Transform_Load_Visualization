```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

dataset = pd.read_excel('matplot_data.xlsx')
dataset.head()
```

```python
plt.scatter(dataset['population'], dataset['salary'], color='blue')
plt.title("Population VS Salary")
plt.xlabel('Population')
plt.ylabel('Salary')
num = np.polyfit(dataset['population'], dataset['salary'], 1)
np.polyval(num, dataset['population'])

plt.show()
```
